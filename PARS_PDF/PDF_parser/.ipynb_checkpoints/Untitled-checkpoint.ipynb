# Работаем с pdf файлами

```python
def convert_pdf_to_image(pdf_path, output_folder, dpi=200):
    # Получить имя файла без расширения
    base_name = os.path.splitext(os.path.basename(pdf_path))[0]

    # Конвертировать PDF в список изображений
    images = convert_from_path(pdf_path, dpi)

    # Если PDF состоит из одной страницы, сохранить с тем же именем файла
    if len(images) == 1:
        image_path = f"{output_folder}/{base_name}.jpg"
        images[0].save(image_path, 'JPEG')
    else:
        # Если PDF содержит несколько страниц, добавить номер страницы к имени файла
        for i, image in enumerate(images):
            image_path = f"{output_folder}/{base_name}_page_{i+1}.jpg"
            image.save(image_path, 'JPEG')

    print(f"Конвертация '{base_name}' завершена. Изображения сохранены в {output_folder}")

# Путь к папке с PDF файлами и папке для сохранения изображений
pdf_folder = 'PDF'
output_folder = 'Output_Images'

# Создать папку для выходных изображений, если она еще не существует
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Получить список всех PDF файлов в папке
pdf_files = [f for f in os.listdir(pdf_folder) if f.endswith('.pdf')]

# Конвертировать каждый PDF файл в изображение
for pdf_file in pdf_files:
    pdf_path = os.path.join(pdf_folder, pdf_file)
    convert_pdf_to_image(pdf_path, output_folder)
```

```python
# Функция улучшения изображения
def enhance_image(file_path):
    # Открыть изображение
    img = Image.open(file_path)

    # Увеличение разрешения
    img = img.resize((img.width * 2, img.height * 2), Image.LANCZOS)

    # Преобразование в градации серого
    img = img.convert('L')

    # Увеличение контрастности
    enhancer = ImageEnhance.Contrast(img)
    img = enhancer.enhance(2)

    # Бинаризация изображения
    img = img.point(lambda x: 0 if x < 128 else 255, '1')

    # Удаление шума
    img = img.filter(ImageFilter.MedianFilter())

    # Сохранение улучшенного изображения вместо исходного
    img.save(file_path)

# Применение функции улучшения ко всем изображениям в папке
for file_name in os.listdir(output_folder):
    if file_name.lower().endswith('.jpg'):  # Проверка, что файл является изображением
        file_path = os.path.join(output_folder, file_name)
        enhance_image(file_path)  # Применение функции улучшения
```

```python
def extract_text_from_images(folder_path, text_folder_path):
    pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

    # Создать папку для текстовых файлов, если она еще не существует
    if not os.path.exists(text_folder_path):
        os.makedirs(text_folder_path)

    for image_file in os.listdir(folder_path):
        if image_file.endswith('.jpg'):
            image_path = os.path.join(folder_path, image_file)
            image = Image.open(image_path)

            # Извлечение текста с помощью OCR
            text = pytesseract.image_to_string(image, lang='rus')

            # Имя файла для сохранения текста
            text_file_name = os.path.splitext(image_file)[0] + '.txt'
            text_file_path = os.path.join(text_folder_path, text_file_name)

            # Сохранение текста в файл
            with open(text_file_path, 'w', encoding='utf-8') as file:
                file.write(text)
            
            print(f"Текст из {image_file} сохранен в {text_file_path}")

# Путь к папке с изображениями и папке для текстовых файлов
image_folder = 'Output_Images'
text_folder = 'TXT'

extract_text_from_images(image_folder, text_folder)
```

# Отдельный блок функций для улучшения текста!

```python
def remove_empty_lines(directory):
    # Проходим по всем файлам в директории
    for filename in os.listdir(directory):
        # Проверяем, является ли файл текстовым
        if filename.endswith('.txt'):
            file_path = os.path.join(directory, filename)
            
            # Читаем файл и удаляем пустые строки
            with open(file_path, 'r', encoding='utf-8') as file:
                lines = [line for line in file if line.strip()]

            # Перезаписываем файл с удаленными пустыми строками
            with open(file_path, 'w', encoding='utf-8') as file:
                file.writelines(lines)
            print(f"Обработан файл: {filename}")

# Вызываем функцию для удаления пустых строк
remove_empty_lines(text_folder)
```

Распарсить нужно следующую информацию:

- Дата;
- Сумма;
- Номер платежного поручения;
- Плательщик (наименование);
- Плательщик ИНН;
- Получатель (наименование);
- Получатель ИНН.

```python
def read_text_files(folder_path):
    texts = {}
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.txt'):
            with open(os.path.join(folder_path, file_name), 'r', encoding='utf-8') as file:
                texts[file_name] = file.read()
    return texts
```

```python
# Функция поиска даты
def find_date(text):
    match = re.search(r'\d{2}\.\d{2}\.\d{4}', text)
    return match.group(0) if match else None

# Функция поиска Платежного поручения
def find_payment_order_number(text):
    match = re.search(r'ПЛАТ[ЕЁ|ЁЕ]+ЖНОЕ ПОРУЧЕНИЕМ?\s*(№|N|M|М|K|К)?\s*[-—]?\s*(\d+)', text)
    return match.group(2) if match else None

# Функция для преобразования числительных, записанных словами, в числа
def word_to_number(words):
    number_dict = {
        'ноль': 0, 'один': 1, 'одна':1, 'два': 2, 'две': 2, 'три': 3, 'четыре': 4,
        'пять': 5, 'шесть': 6, 'семь': 7, 'восемь': 8, 'девять': 9,
        'десять': 10, 'одиннадцать': 11, 'двенадцать': 12,
        'тринадцать': 13, 'четырнадцать': 14, 'пятнадцать': 15,
        'шестнадцать': 16, 'семнадцать': 17, 'восемнадцать': 18,
        'девятнадцать': 19, 'двадцать': 20, 'тридцать': 30,
        'сорок': 40, 'пятьдесят': 50, 'шестьдесят': 60,
        'семьдесят': 70, 'восемьдесят': 80, 'девяносто': 90,
        'сто': 100, 'двести': 200, 'триста': 300, 'четыреста': 400,
        'пятьсот': 500, 'шестьсот': 600, 'семьсот': 700,
        'восемьсот': 800, 'девятьсот': 900
    }
    multiplier_dict = {
        'тысяча': 1000, 'тысячи': 1000, 'тысяч': 1000,
        'миллион': 1000000, 'миллиона': 1000000, 'миллионов': 1000000
    }

    words = words.split()
    total_sum = 0
    current_sum = 0

    for word in words:
        if word in multiplier_dict:
            total_sum += current_sum * multiplier_dict[word]
            current_sum = 0
        else:
            current_sum += number_dict.get(word, 0)

    total_sum += current_sum
    return total_sum

# Функция поиска и преобразования суммы
def find_and_convert_sum(text):
    text = text.lower()
    # Обновленное регулярное выражение, учитывающее возможное отсутствие копеек
    match = re.search(r'(сумма\s+прописью|сумма|прописью)\s+([^\d]+)\s+рубл[еёй]+(?:\s+(\d+)\s+копе[йкйеки]+)?', text)
    if match:
        sum_in_words = match.group(2)
        sum_in_kopecks = match.group(3) or '00'
        sum_in_numbers = word_to_number(sum_in_words)

        return f"{sum_in_numbers}-{sum_in_kopecks.zfill(2)}"
    else:
        return None

# Функция для поиска ИНН плательщика/получателя
def find_inns(text):
    # Приведение текста к нижнему регистру для учета разных вариантов написания "ИНН"
    text = text.lower()
    # Регулярное выражение теперь учитывает "инн", за которым может следовать тире или дефис
    matches = re.findall(r'инн\s*[-—_]?\s*(\d{10,12})', text)
    payer_inn = matches[0] if len(matches) > 0 else None
    receiver_inn = matches[1] if len(matches) > 1 else None
    return payer_inn, receiver_inn

def find_payer_name(text):
    text = text.lower()
    # Извлечение строк, следующих за "ИНН"
    matches = re.findall(r'инн[^\n]*\n([^\n]+)', text)
    # Возвращение наименования, следующего за первым "ИНН"
    return matches[0].strip() if matches else None

def find_receiver_name(text):
    text = text.lower()
    matches = re.findall(r'инн[^\n]*\n([^\n]+)', text)
    # Возвращение наименования, следующего за вторым "ИНН"
    return matches[1].strip() if len(matches) > 1 else None

# Добавляем вызов этой функции в extract_and_save_information
def extract_and_save_information(texts, output_file=None):
    # Создание списка для хранения извлеченной информации
    data = []

    for file_name, text in texts.items():
        date = find_date(text)
        payment_order = find_payment_order_number(text)
        sum_in_numbers = find_and_convert_sum(text)
        payer_inn, receiver_inn = find_inns(text)
        payer_name = find_payer_name(text)
        receiver_name = find_receiver_name(text)

        # Добавление информации в список
        data.append({
            'Имя файла': file_name,
            'Дата': date,
            'Номер платежного поручения': payment_order,
            'Сумма': sum_in_numbers,
            'Плательщик': payer_name,
            'ИНН плательщика': payer_inn,
            'Получатель': receiver_name,
            'ИНН получателя': receiver_inn
        })

    # Преобразование списка в DataFrame
    df = pd.DataFrame(data)

    # Сохранение DataFrame в CSV файл
    if output_file:
        df.to_csv(output_file, index=False, encoding='utf-8')
    return df

# Пример использования
texts = read_text_files('TXT')
df = extract_and_save_information(texts, 'Processed_TXT.csv')
df
```
