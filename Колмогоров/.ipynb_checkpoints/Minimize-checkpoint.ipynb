```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.optimize import minimize

def fi(x1, x2, i):
    return (x1 - i)**2 + (x2 + i)**2

def system(x):
    return np.sum([fi(x[0], x[1], i) for i in range(1, 12)])

# Оптимизация каждого уравнения по отдельности
optimized_values = []
for i in range(1, 12):
    result = minimize(lambda x: fi(x[0], x[1], i), [0, 0])
    optimized_values.append(result.x)

# Подбор оптимальной пары (x1, x2) из найденных точек
optimal_solution = None
best_f = float('inf')
for values in optimized_values:
    f_value = system(values)
    if f_value < best_f:
        best_f = f_value
        optimal_solution = values

# Вывод результатов 
print("Минимум функции fi(x1,x2)=(x1-i)^2+(x2+i)^2 :")
print("   i          x1        x2        f(x1, x2)")
for i, value in enumerate(optimized_values, start=1):
    f_value = fi(value[0], value[1], i)
    print(f"  {i:2d}     {value[0]:10.4f} {value[1]:10.4f} {f_value:13.4e}")

optimal_f = system(optimal_solution)
print("\nМинимум системы:")
print(f"x1={optimal_solution[0]:.4f}, x2={optimal_solution[1]:.4f}, f={optimal_f:.4e}")

# Построение графика
x1 = np.linspace(-10, 20, 100)
x2 = np.linspace(-20, 10, 100)
X1, X2 = np.meshgrid(x1, x2)
Z = np.zeros_like(X1)

for i in range(Z.shape[0]):
    for j in range(Z.shape[1]):
        Z[i, j] = system([X1[i, j], X2[i, j]])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, X2, Z, cmap='viridis')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('f(x1, x2)')

# Добавление оптимальной точки на график
ax.scatter(optimal_solution[0], optimal_solution[1], optimal_f, color='red', label='Минимум')
ax.legend()

plt.title('График системы')
plt.show()
```
