```python
import numpy as np
import matplotlib.pyplot as plt

def generate_matrix(N, α, var_diag=1, var_off_diag=1):
    if 0 <= α <= 1:  # Гауссов унитарный ансамбль (GUE)
        A = np.random.normal(0, np.sqrt(var_diag), size=(N, N))  # Диагональные элементы
        A = (A + A.T) / 2  # делаем матрицу симметричной
        return A
    else:
        raise ValueError("Значение α должно быть в диапазоне от 0 до 1.")

def V(ensemble, alpha, beta):
#     if alpha != 1:
#         raise ValueError("Функция V применима только когда alpha = 1")
    N = ensemble.shape[0]
    eigenvalues = np.linalg.eigvals(ensemble)
    term1 = alpha**2 * np.sum(eigenvalues**2)
    term2 = (1 - alpha**2) * np.sum((eigenvalues[i] - eigenvalues[j])**2 for i in range(N) for j in range(i+1, N))
    return term1 + term2

def wigner_semicircle(x, v, h):
    return np.sqrt((4*v - x**2) / (2 * np.pi * v * h)) if abs(x) <= 2 * np.sqrt(v * h) else 0

N = 1000
alpha = 0
ensemble = generate_matrix(N, alpha)
result = V(ensemble, alpha, beta=2)
print("V(λ) = ", result)

eigenvalues = np.linalg.eigvals(ensemble)
eigenvalues = eigenvalues / np.sqrt(N)
v = np.var(eigenvalues)

# Уменьшаем высоту полукруга, оставив ширину неизменной
h = 3.2# Коэффициент для управления высотой полукруга

x = np.linspace(-2*np.sqrt(v), 2*np.sqrt(v), 1000)
y = np.array([wigner_semicircle(xi, v, h) for xi in x])

plt.hist(eigenvalues, bins=50, density=True, alpha=0.5, label='Собственные значения')
plt.plot(x, y, label='Полукруговой закон Вигнера')
plt.title('Распределение собственных значений в гауссовом унитарном ансамбле')
plt.xlabel('Собственные значения')
plt.ylabel('Плотность вероятности')
plt.legend(loc='upper right')
plt.show()
```

В этой модификации изменено выражение. Теперь чем ближе alpha к 0, тем больше влияние вносит компонента B, что приводит к большей асимметрии в ансамбле матриц. А при alpha близком к 1, влияние компоненты B уменьшается, и ансамбль становится более похожим на распределение Вигнера.

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_matrix(N, alpha, var_diag=1, var_off_diag=1):
    if 0 <= alpha <= 1:  # Более асимметричная матрица
        A = np.random.normal(0, np.sqrt(var_diag), size=(N, N))  # Диагональные элементы
        B = np.random.normal(0, np.sqrt(var_off_diag), size=(N, N))  # Недиагональные элементы
        A = (A + A.T) / 2  # Делаем матрицу симметричной
        A += alpha * B  # Добавляем недиагональные элементы с коэффициентом alpha
        return A
    else:  
        raise ValueError("Значение α должно быть в диапазоне от 0 до 1.")

def V(ensemble, alpha, beta):
    N = ensemble.shape[0]
    eigenvalues = np.linalg.eigvals(ensemble)
    term1 = alpha**2 * np.sum(eigenvalues**2)
    term2 = (1 - alpha**2) * np.sum((eigenvalues[i] - eigenvalues[j])**2 for i in range(N) for j in range(i+1, N))
    return term1 + term2

def wigner_semicircle(x, v, h):
    return np.sqrt((4*v - x**2) / (2 * np.pi * v * h))

N = 1000
alphas = [0, 0.01, 0.1, 0.5, 1]  # Пример значений alpha для перебора

for alpha in alphas:
    ensemble = generate_matrix(N, alpha)
    result = V(ensemble, alpha, beta=2)
    print("V(λ) =", result)

    eigenvalues = np.linalg.eigvals(ensemble)
    eigenvalues = eigenvalues / np.sqrt(N)

    hist, bins = np.histogram(eigenvalues, bins=50, density=True)
    bin_centers = (bins[:-1] + bins[1:]) / 2

    v = np.var(eigenvalues)
    h = 3.2

    x = np.linspace(-2.5*np.sqrt(v), 2.5*np.sqrt(v), 1000)
    y = wigner_semicircle(x, v, h)

    plt.figure()  # Создаем новую фигуру для каждого значения alpha
    plt.hist(eigenvalues, bins=50, density=True, alpha=0.5, label='Собственные значения')
    plt.plot(x, y, label='Полукруговой закон Вигнера')
    plt.title(f'Распределение собственных значений (alpha={alpha}) в асимметричных вещественных матрицах')
    plt.xlabel('Собственные значения')
    plt.ylabel('Плотность вероятности')
    plt.legend(loc='upper right')

plt.show()  # Отображаем все фигуры с гистограммами
```
