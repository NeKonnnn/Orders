```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
```

```python
# Загрузка и вывод первых строк файла Excel для ознакомления со структурой данных
data = pd.read_excel('cl.xlsx')

# Показать первые строки таблицы для ознакомления
data.head()
```

Таблица содержит 29 столбцов, включая неименованный столбец, который, возможно, является индексом, столбец 'Region' с названиями регионов, столбец 'Y' с годом (все строки содержат 2019 год), столбцы 'z1' до 'z26' с различными показателями для нормализации, и столбец 'население/Наркоманы' с некоторыми данными, которые, как вы указали, не следует учитывать при кластеризации.

Перед нормализацией данных, удалим неиспользуемые столбцы и преобразуем оставшиеся показатели 
$z_1, z_2,..., z_26$ так, чтобы они имели среднее значение 0 и стандартное отклонение 1. После этого можно будет приступить к кластеризации.​

```python
# Удаление неиспользуемых столбцов
data_for_clustering = data.drop(['Unnamed: 0', 'Region', 'Y', 'население/Наркоманы'], axis=1)

# Инициализация нормализатора
scaler = StandardScaler()

# Нормализация данных (пронормированные данные будут иметь среднее значение 0 и стандартное отклонение 1)
normalized_data = scaler.fit_transform(data_for_clustering)

# Преобразование обратно в DataFrame для удобства
normalized_data_df = pd.DataFrame(normalized_data, columns=data_for_clustering.columns)

# Вывод первых пяти строк нормализованных данных
normalized_data_df.head()
```

Показатели $z_1, z_2,..., z_26$ были успешно пронормированы. Теперь мы можем перейти к кластеризации данных.

Сначала применим иерархическую кластеризацию. Для этого мы используем метод AgglomerativeClustering из библиотеки scikit-learn. Затем применим метод k-средних, используя класс KMeans. Количество кластеров выберем на основе метода локтя или силуэта, если необходимо.

```python
# Определение функции для вычисления оптимального количества кластеров с использованием метода локтя
def optimal_number_of_clusters(data):
    from scipy.cluster.hierarchy import dendrogram, linkage
    from matplotlib import pyplot as plt
    
    # Вычисление связей между данными для дендрограммы
    linked = linkage(data, 'ward')

    # Построение дендрограммы
    plt.figure(figsize=(10, 7))
    dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
    plt.show()

# Вызов функции для построения дендрограммы и определения оптимального числа кластеров
optimal_number_of_clusters(normalized_data_df)
```

Дендрограмма показывает иерархическую структуру возможных кластеров. Оптимальное количество кластеров можно выбрать, обрезав дендрограмму на разной высоте и выбирая количество вертикальных линий, которые пересекаются с этой горизонтальной линией.

Теперь, исходя из дендрограммы, выберем количество кластеров для иерархической кластеризации. Визуально можно предположить, что разумное количество кластеров будет между 3 и 5, так как на дендрограмме видны крупные "ветви", которые можно интерпретировать как отдельные кластеры. Чтобы быть более точными, мы можем выбрать количество кластеров, где расстояние между соседними соединениями становится значительно больше, что указывает на объединение отдельных, хорошо отделенных кластеров.

Для демонстрации выберем 4 кластера и выполним иерархическую кластеризацию с этим количеством.

```python
# Применение иерархической кластеризации с выбранным числом кластеров
n_clusters = 4
hierarchical_clustering = AgglomerativeClustering(n_clusters=n_clusters)
hierarchical_labels = hierarchical_clustering.fit_predict(normalized_data_df)

# Добавим метки кластера в исходный DataFrame для удобства анализа
data['Hierarchical Cluster'] = hierarchical_labels

# Просмотрим результаты иерархической кластеризации
data[['Region', 'Hierarchical Cluster']].head(10)
```

Иерархическая кластеризация была выполнена, и каждому региону был присвоен номер кластера от 0 до 3. Теперь мы можем перейти к методу k-средних.

Прежде чем применять метод k-средних, полезно определить оптимальное количество кластеров. Для этого мы можем использовать метод локтя, который включает в себя вычисление суммы квадратов расстояний от каждой точки до центра её кластера для разного количества кластеров и выбор количества кластеров, при котором это значение начинает уменьшаться намного медленнее (точка "изгиба").

Построим график метода локтя для наших данных.

```python
# Функция для вычисления суммы квадратов внутрикластерных расстояний
def calculate_wcss(data):
    wcss = []
    for n in range(1, 11):
        kmeans = KMeans(n_clusters=n, init='k-means++', max_iter=300, n_init=10, random_state=42)
        kmeans.fit(data)
        wcss.append(kmeans.inertia_)
    return wcss

# Вычисление суммы квадратов расстояний для разного количества кластеров
wcss = calculate_wcss(normalized_data_df)

# Построение графика метода локтя
plt.figure(figsize=(10, 8))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Метод локтя для определения оптимального количества кластеров')
plt.xlabel('Количество кластеров')
plt.ylabel('Сумма квадратов внутрикластерных расстояний')
plt.xticks(range(1, 11))
plt.show()
```

На графике метода локтя видно, что сумма квадратов расстояний резко уменьшается до 4 кластеров и затем снижается более плавно. Это указывает на то, что оптимальное количество кластеров для метода k-средних тоже может быть 4.

Теперь применим метод k-средних с 4 кластерами к нашим данным.

```python
# Применение метода k-средних с оптимальным числом кластеров
kmeans = KMeans(n_clusters=n_clusters, init='k-means++', max_iter=300, n_init=10, random_state=42)
kmeans_labels = kmeans.fit_predict(normalized_data_df)

# Добавление меток кластера от k-средних к исходному DataFrame
data['KMeans Cluster'] = kmeans_labels

# Просмотрим результаты метода k-средних
data[['Region', 'KMeans Cluster']].head(10)  
```

Метод k-средних также был применён, и каждому региону был присвоен номер кластера от 0 до 3.

Теперь у нас есть результаты кластеризации для каждого региона, полученные двумя разными методами: иерархической кластеризацией и методом k-средних.

```python
# Создание объекта Excel writer
with pd.ExcelWriter('clustering_results_separate_sheets.xlsx') as writer:
    # Сохранение результатов иерархической кластеризации на отдельный лист
    data[['Region', 'Hierarchical Cluster']].to_excel(writer, index=False, sheet_name='Иерархическая')
    # Сохранение результатов метода k-средних на отдельный лист
    data[['Region', 'KMeans Cluster']].to_excel(writer, index=False, sheet_name='K-средних')
```
