**1.b.** Постройте следующий график:

- процентое изменение цены за акцию Tesla за день по отношению к цене на открытии торгов.

На оси **x** должны быть отмечены даты из data.index. Значения оси **y** должны отражать, на сколько процентов увеличилась (положительное число) или на сколько процентов упала (отрицательное число) цена акции на закрытии Close по сравнению с ценой открытия Open.

В какой день наблюдалось самое резкое понижение цены за акцию? А самое большое повышение? На сколько процентов?

*Подсказка: создайте новую колонку Diff, в которой посчитайте изменение акции за день в процентах*

**Ваш ответ:**

Для начала создадим новую колонку Diff, в которой рассчитаем процентное изменение цены акции за день (от Open до Close).
$$\text{Diff} = \left( \frac{\text{Close} - \text{Open}}{\text{Open}} \right) \times 100$$
Далее построим график и найдем дни с максимальным повышением и понижением цены.

```python
# Рассчитываем процентное изменение цены акции за день
data['Diff'] = ((data['Close'] - data['Open']) / data['Open']) * 100

# Построение графика
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(data['Date'], data['Diff'], label='Процентное изменение', color='green')

# Форматирование графика
ax.set_title('Процентное изменение цены акции Tesla за день относительно цены на открытии торгов')
ax.set_xlabel('Дата')
ax.set_ylabel('Процентное изменение')
ax.axhline(0, color='black', linewidth=0.5)  # линия для нулевого изменения
ax.xaxis.set_major_locator(mdates.MonthLocator())  # установка мажорных делений по месяцам
ax.xaxis.set_major_formatter(DateFormatter("%b"))  # установка формата месяца

# Отображение графика
plt.grid(True)
plt.show()
```

На графике представлено процентное изменение цены акции Tesla за день относительно цены на открытии торгов.

```python
# Находим день с максимальным повышением цены
max_increase_day = data.loc[data['Diff'].idxmax()]
# Находим день с максимальным понижением цены
max_decrease_day = data.loc[data['Diff'].idxmin()]

max_increase_day['Date'], max_increase_day['Diff'], max_decrease_day['Date'], max_decrease_day['Diff']
```

- Самое большое повышение цены акции было 3 февраля 2020 года и составило приблизительно 15.78%.
- Самое резкое понижение цены акции произошло 5 февраля 2020 года и составило приблизительно −10.76%.

**1.с.** С помощью **pandas** преобразуйте индекс датасета в формат «год-месяц» и сгруппируйте данные так, чтобы новому индексу соответствовали средние значения по колонкам в указанном месяце и годе.

Напишите месяц, когда объем акций впервые превысил 60M. Какая средняя цена акции на закрытии торгов соответствует этому месяцу?

*Подсказка: используйте параметр hover_data.*

**Ваш ответ:**

```python
# Преобразование индекса датасета в формат "год-месяц"
data['Year_Month'] = data['Date'].dt.to_period('M')

# Группировка данных по новому индексу и расчет средних значений по колонкам
grouped_data = data.groupby('Year_Month').mean()

# Поиск месяца, когда объем акций впервые превысил 60M
first_month_above_60M = grouped_data[grouped_data['Volume'] > 60000000].index.min()

# Определение средней цены акции на закрытии торгов для этого месяца
average_close_price = grouped_data.loc[first_month_above_60M, 'Close']

first_month_above_60M, average_close_price
```

Объем акций впервые превысил 60M в январе 2020 года. Средняя цена акции на закрытии торгов в этот месяц составила примерно $35.24.

**1.a.** Загрузите данные. Постройте график scatter зависимости параметра ram от battery_power. С помощью одного графика ответьте на следующие вопросы:

*Подсказка: переведите значения price_range в строки для того, чтобы легенда корректно отображалась;
используйте параметры color, size*.

**Ваш ответ:**

Построим диаграмму рассеяния (scatter plot) для зависимости параметра ram от battery_power, используя разные цвета для различных диапазонов цен (price_range) и размер точек в зависимости от времени разговора (talk_time). Это поможет нам визуально оценить зависимости между этими параметрами.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Конвертация price_range в строки для корректного отображения в легенде
data['price_range_str'] = data['price_range'].astype(str)

# Построение scatter plot
plt.figure(figsize=(12, 8))
sns.scatterplot(data=data, x='battery_power', y='ram', hue='price_range_str', size='talk_time', palette="deep", alpha=0.6, edgecolor=None)

# Добавление заголовка и меток осей
plt.title('Зависимость RAM от мощности батареи с учетом ценового диапазона и времени разговора')
plt.xlabel('Мощность батареи (battery_power)')
plt.ylabel('RAM')

# Отображение легенды
plt.legend(loc='upper left', title='Price Range / Talk Time')

# Отображение графика
plt.show()
```

На графике выше каждая точка представляет собой телефон, где ось X обозначает мощность батареи (battery_power), ось Y - объем RAM (ram). Цвет точек обозначает ценовой диапазон (price_range), а размер точек представляет время разговора (talk_time).

**Анализ графика**:
1. Зависимость ram от battery_power:
- На первый взгляд, зависимость между ram и battery_power не наблюдается.
2. Влияние battery_power на стоимость телефона:
- Точки с более высокой мощностью батареи не сконцентрированы вокруг высоких значений price_range, поэтому можно сказать, что нет прямой зависимости между мощностью батареи и стоимостью телефона.
3. Влияние ram на стоимость телефона:
- Точки с более высоким значением ram чаще имеют более высокий price_range, что указывает на то, что больше ram обычно означает более высокую стоимость телефона.
4. Зависимость talk_time от battery_power и ram:
- Нет явной зависимости talk_time от battery_power и ram, так как точки с разными размерами равномерно распределены по всему графику.

**Выводы**:
- Нет явной зависимости ram от battery_power.
- Большее значение battery_power не гарантирует более высокую стоимость телефона.
- Больший объем ram обычно связан с более высокой стоимостью телефона.
- Нет явной зависимости talk_time от battery_power и ram.
