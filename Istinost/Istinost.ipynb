**Задание**: 
Формулы задаются таблицей истинности

Требуется для введенного набора формул определить классы поста каждой и сделать вердикт, является ли набор полным
Шкала оценивания: всего 10 баллов

- 0-6 баллов за работоспособность кода
- 0-1 баллов за возможность вводить входные данные (формулы меняются без редактирования кода)
- 0-1 баллов за подсказки при вводе
- 0-1 баллов за написанную документацию в которой указывается как все работает
- 0-1 баллов за защиту от дурака (например отрицательные числа вместо положительных на вводе).

**Классы булевых функций**:
- Класс Т_0 - класс функций, сохраняющих 0
- Класс Т_1 - класс функций, сохраняющих 1
- Класс М - класс монотонных функций
- Класс S - класс самодвойственных функций
- Класс L - класс линейных функций

```python
from itertools import product
from typing import List, Tuple

def check_T0(table: List[int]) -> bool:
    """ Проверка на принадлежность к классу Т_0 (сохраняющий 0). """
    return table[0] == 0

def check_T1(table: List[int], n_vars: int) -> bool:
    """ Проверка на принадлежность к классу Т_1 (сохраняющий 1). """
    return table[-1] == 1

def check_M(table: List[int], n_vars: int) -> bool:
    """ Проверка на принадлежность к классу М (монотонность). """
    for i in range(len(table) - 1):
        for j in range(i + 1, len(table)):
            binary_i = format(i, f'0{n_vars}b')
            binary_j = format(j, f'0{n_vars}b')
            if all(bit_i <= bit_j for bit_i, bit_j in zip(binary_i, binary_j)):
                if table[i] > table[j]:
                    return False
    return True

def check_S(table: List[int], n_vars: int) -> bool:
    """ Проверка на принадлежность к классу S (самодвойственность). """
    for i in range(len(table) // 2):
        if table[i] == table[-1 - i]:
            return False
    return True

def check_L(table: List[int], n_vars: int) -> bool:
    """ Проверка на принадлежность к классу L (линейность). """
    for combination in product([0, 1], repeat=n_vars):
        sum_comb = sum(combination) % 2
        index = int("".join(map(str, combination)), 2)
        if table[index] != sum_comb:
            return False
    return True

def read_function():
    """ Считывает булеву функцию от пользователя. """
    while True:
        try:
            n = int(input("Введите количество переменных для функции: "))
            if n < 0:
                raise ValueError
            break
        except ValueError:
            print("Пожалуйста, введите положительное целое число.")

    while True:
        try:
            f = list(map(int, input(f"Введите таблицу истинности (как последовательность из 0 и 1, длиной {2 ** n}): ").split()))
            if len(f) != 2 ** n or any(v not in [0, 1] for v in f):
                raise ValueError
            break
        except ValueError:
            print(f"Пожалуйста, введите последовательность из {2 ** n} нулей и единиц.")

    return f, n

# def main():
#     functions = []
#     post_classes = {"T0": [], "T1": [], "M": [], "S": [], "L": []}

#     while True:
#         f, n = read_function()
#         functions.append((f, n))

#         # Check and assign Post classes
#         if check_T0(f):
#             post_classes["T0"].append(f)
#         if check_T1(f, n):
#             post_classes["T1"].append(f)
#         if check_M(f, n):
#             post_classes["M"].append(f)
#         if check_S(f, n):
#             post_classes["S"].append(f)
#         if check_L(f, n):
#             post_classes["L"].append(f)

#         if input("Желаете добавить еще функцию? (да/нет): ").lower() != "да":
#             break

#     # Check for completeness
#     is_complete = all([post_classes[cls] for cls in post_classes])

#     # Вывод результатов
#     print("\nКлассы Поста:")
#     for cls in post_classes:
#         print(f"{cls}: {len(post_classes[cls])} функций")

#     print(f"\nЯвляется ли набор функций полным? {'Да' if is_complete else 'Нет'}")

def main():
    functions = []
    # Изменяем структуру для хранения принадлежности каждой функции к каждому классу
    post_classes = {"T0": [], "T1": [], "M": [], "S": [], "L": []}

    while True:
        f, n = read_function()
        functions.append((f, n))

        # Check and assign Post classes
        post_classes["T0"].append(1 if check_T0(f) else 0)
        post_classes["T1"].append(1 if check_T1(f, n) else 0)
        post_classes["M"].append(1 if check_M(f, n) else 0)
        post_classes["S"].append(1 if check_S(f, n) else 0)
        post_classes["L"].append(1 if check_L(f, n) else 0)

        if input("Желаете добавить еще функцию? (да/нет): ").lower() != "да":
            break

    # Check for completeness
    is_complete = all([any(post_classes[cls]) for cls in post_classes])

    # Вывод результатов
    print("\nПринадлежность к классам Поста:")
    for cls in post_classes:
        print(f"{cls}: {post_classes[cls]}")

    print(f"\nЯвляется ли набор функций полным? {'Да' if is_complete else 'Нет'}")

main()
```

**Общее описание**
Код позволяет пользователю вводить булевы функции через их таблицы истинности и определяет, к каким классам Поста принадлежит каждая функция. Классы Поста включают функции, сохраняющие 0 (T0), сохраняющие 1 (T1), монотонные (M), самодвойственные (S) и линейные (L). После ввода всех функций код проверяет, является ли введенный набор функций полным, то есть способным выразить любую другую булеву функцию.

**Функции**
1. `check_T0(table: List[int]) -> bool`

- Проверяет, сохраняет ли функция 0 (принадлежность к классу Т0).
- Возвращает True, если первый элемент таблицы истинности равен 0.

2. `check_T1(table: List[int], n_vars: int) -> bool`

- Проверяет, сохраняет ли функция 1 (принадлежность к классу Т1).
- Возвращает True, если последний элемент таблицы истинности равен 1.

3. `check_M(table: List[int], n_vars: int) -> bool`

- Проверяет монотонность функции (принадлежность к классу М).
- Сравнивает значения таблицы истинности для всех пар входных комбинаций, где одна комбинация не меньше другой.

4. `check_S(table: List[int], n_vars: int) -> bool`

- Проверяет самодвойственность функции (принадлежность к классу S).
- Проверяет, что значения функции для каждой пары дополнительных наборов входных переменных противоположны.

5. `check_L(table: List[int], n_vars: int) -> bool`

- Проверяет линейность функции (принадлежность к классу L).
- Проверяет, соответствует ли результат функции сумме её аргументов по модулю 2 для всех возможных комбинаций входных значений.
`
6. `read_function() -> Tuple[List[int], int]`

- Считывает от пользователя количество переменных и таблицу истинности булевой функции.
- Обрабатывает ввод, проверяя корректность данных.

7. `main()`

- Основная функция, которая управляет процессом ввода функций и их анализом.
- Сохраняет введенные функции и проверяет их на принадлежность к классам Поста.
- Определяет, является ли набор функций полным.

**Использование**
Пользователь запускает функцию main(), после чего вводит количество переменных и таблицу истинности для каждой булевой функции. После ввода каждой функции код автоматически определяет её принадлежность к классам Поста. После ввода всех функций код выводит статистику по классам и сообщает, является ли набор функций полным.

**Примечания**
Защита от некорректного ввода реализована в функции read_function.
Проверка на полноту основывается на наличии хотя бы одной функции из каждого класса Поста.
Код предназначен для работы с булевыми функциями, введенными в виде таблиц истинности.
